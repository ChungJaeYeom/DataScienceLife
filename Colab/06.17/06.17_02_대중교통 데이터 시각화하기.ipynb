무임승차 비율이 가장 높은 역은 어디이까?

유임승차 / 무임승차 비율의 최댓값 찾기 -> 기존 0 데이터로 되어 있음 <br>
rate = 유임승차 인원 / 무임 승차 인원


```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = 0 # 최대 유임승차 비율
rate = 0 # 유임승차 비율

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
  
  if row[6] != 0:
    rate = row[4] / row[6] # 유임승차 / 무임승차

  if rate > mx:
    mx = rate
    
print(mx)
```

비율 계산 방식 변경 rate = 유임승차 인원 / 전체(유임+무임) 인원


```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = 0 # 최대 유임승차 비율
rate = 0 # 유임승차 비율

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
  
  if row[6] != 0:
    rate = row[4] / (row[4] + row[6]) # 유임승차 / 전체(유임+무임)

  if rate > mx:
    mx = rate
    mx_station = row[3] + ' ' + row[1] # 지하철 역 + 호선 정보 ex) 서울역 1호선

print(mx_station, round(mx*100,2),'%')
```

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = 0 # 최대 유임승차 비율
rate = 0 # 유임승차 비율
mx_station = '' #최대 유임승차 역 정보

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
  
  if row[6] != 0 and (row[4] + row[6]) > 100000:  #유무임 승차 인원 합해서 100,000명 이상인 경우만
    rate = row[4] / (row[4] + row[6]) # 유임승차 / 전체(유임+무임)

  if rate > mx:
    mx = rate
    mx_station = row[3] + ' ' + row[1] # 지하철 역 + 호선 정보 ex) 서울역 1호선

print(mx_station, round(mx*100,2),'%')
```

유무임 승하차 인원이 가장 많은 역 찾기

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = [0] * 4 #최대 유임승차, 유임하차, 무임승차, 무임하차 인원수
mx_station = [''] * 4 #최대 유임승차, 유임하차, 무임승차, 무임하차 역 정보 

label = ['유임승차', '유임하차', '무임승차', '무임하차']

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
    if row[i] > mx[i-4]:
      mx[i-4] = row[i]
      mx_station[i-4] = row[3] + ' ' + row[1]

for i in range(4):
  print(label[i]+':'+mx_station[i],mx[i])
```

최소 찾기

```python
import csv
f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = [1000000000]*4 #최대 유임승차, 유임하차, 무임승차, 무임하차 인원수
mx_station = [''] * 4 #최대 유임승차, 유임하차, 무임승차, 무임하차 역 정보 

label = ['유임승차', '유임하차', '무임승차', '무임하차']

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])

    if row[i] > 0: #if row[i] != 0: 문자열이 아니니까 비교연산자를 나와야함
      if row[i] <= mx[i-4]:
        mx[i-4] = row[i]
        mx_station[i-4] = row[3] + ' ' + row[1]

for i in range(4):
  print(label[i]+':'+mx_station[i],mx[i])
```

모든 역의 유무임 승차 비율은 다를까? 파이 차트로 나타내기

```python
import csv
import matplotlib.pyplot as plt


f = open('subwayfee.csv')
data = csv.reader(f)
next(data)

mx = [0] * 4 #최대 유임승차, 유임하차, 무임승차, 무임하차 인원수
mx_station = [''] * 4 #최대 유임승차, 유임하차, 무임승차, 무임하차 역 정보 

label = ['유임승차', '유임하차', '무임승차', '무임하차']

for row in data:
  for i in range(4,8):
    row[i] = int(row[i])
    plt.figure(dpi=110)
    plt.title(row[3] + ' ' + row[1])
    plt.pie(row[4:8], labels = label, autopct='%.1f%%')
    plt.axis('equal')
    plt.savefig(row[3] + ' ' + row[1] + '.png')
    plt.show()
```
