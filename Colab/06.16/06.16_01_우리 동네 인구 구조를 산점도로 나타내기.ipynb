꺾은선 그래프로 표현하기

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []

name = input("알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요.: ")

for row in data:
  if name in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',', '')))
    for i in row[106:]: 
      f.append(int(i.replace(',', '')))
    break

plt.style.use('ggplot')
plt.figure(dpi=110) 
plt.rc('font',family='NanumBarunGothic')

# 꺾은선 그래프로 남여 인구수 데이터 표현학
plt.plot(m,label = '남성')
plt.plot(f,label = '여성')
plt.legend()
plt.show()
```

연령별로 남성 인구수에서 여성 인구수를 뺀 차이의 막대그래프 그리기


```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)

result = [] # 남성 인구 - 여성 인구 값 담을 리스트 변수 선언

name = input("알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요.: ")

for row in data:
  if name in row[0]:
    for i in range(3,104):
      result.append(int(row[i].replace(',','')) - int(row[i+103].replace(',','')))
    break

plt.style.use('ggplot')
plt.figure(dpi=110) 
plt.rc('font',family='NanumBarunGothic')
plt.rcParams['axes.unicode_minus'] = False

plt.title(name + ' 지역의 남녀 성별 인구 차이')
plt.bar(range(101),result)
plt.show()
```

위치, 크기가 서로 다른 100개의 점 만들기

```python
import matplotlib.pyplot as plt
import random

x = []
y = []
size = []

for i in range(100):
  x.append(random.randint(50,100))
  y.append(random.randint(50,100))
  size.append(random.randint(10,100))


plt.style.use('ggplot')
plt.figure(dpi=110) 
plt.scatter(x,y,s=size)
plt.show()
```

```python
plt.style.use('ggplot')
plt.figure(dpi=110) 
plt.scatter(x,y,s=size,c=size,cmap='jet')
plt.colorbar()
plt.show()
```

# alpha 투명도 추가 0~1: 0에 가까울수록 투명, 1에 가까울수록 불투명


```python
plt.style.use('ggplot')
plt.figure(dpi=110) 
plt.scatter(x,y,s=size,c=size,cmap='jet',alpha=0.8)
plt.colorbar()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('gender.csv', encoding = 'cp949')
data = csv.reader(f)

m = []
f = []

m1 = []
f1 = []


for row in data:
  if '세종특별자치시' in row[0]:
    for i in row[3:104]:
      m.append(int(i.replace(',', '')))
    for i in row[106:]: 
      f.append(int(i.replace(',', '')))
    break

for row in data:
  if '제주특별자치도' in row[0]:
    for i in row[3:104]:
      m1.append(int(i.replace(',', '')))
    for i in row[106:]: 
      f1.append(int(i.replace(',', '')))
    break

plt.style.use('ggplot')
plt.figure(dpi=160) 
plt.rc('font',family='NanumBarunGothic')
plt.title('세종특별자치시와 제주특별자치도 남성 여성의 인구차이')

# 꺾은선 그래프로 남여 인구수 데이터 표현학
plt.plot(m,label = '남성')
plt.plot(f,label = '여성')
plt.plot(m1,label = '남성')
plt.plot(f1,label = '여성')
plt.legend()
plt.show()
```
