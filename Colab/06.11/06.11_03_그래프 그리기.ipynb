```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('plotting') # 타이틀 속성 추가
plt.plot([1,2],[120.4,0.1]) # (x축,y축) 지정해서 plot()함수에 입력
plt.show() # 그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('plotting') # 타이틀 속성 추가

plt.plot([10,20,30,40],label='asc') # 증가를 의미하는 asc 범례
plt.plot([40,30,20,10],label='desc') # 감소를 의미하는 desc 범례

plt.legend(loc=6) #그래프의 범례 속성 추가 / loc = 6 왼쪽 중앙 위치 설정도 가능
plt.show() # 그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40],color='skyblue',label='sktblue') # 하늘색 그래프
plt.plot([40,30,20,10],'pink',label='pink') # 분홍색 그래프

plt.legend() #그래프의 범례 속성 추가 / loc = 6 왼쪽 중앙 위치 설정도 가능
plt.show() # 그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('color') # 타이틀 속성 추가

plt.plot([10,20,30,40],color='r',label='red') # 기본 색상은 약어로 표시 가능 / b=blue, k=black 구분위해
plt.plot([40,30,20,10],'b',label='blue') 

plt.legend() #그래프의 범례 속성 추가 / loc = 6 왼쪽 중앙 위치 설정도 가능
plt.show() # 그래프 보여주기
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('linestyle') 

plt.plot([10,20,30,40],color='k',linestyle='--',label='dashed') 
plt.plot([40,30,20,10],'g',ls = ':',label='dotted') 

plt.legend() 
plt.show() 
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('marker') 

plt.plot([0,1],'r.',label='circle') #점
plt.plot([0.1,120.4],'g^',label='triangle up') #삼각형 

plt.legend() 
plt.show() 
```

```python
import matplotlib.pyplot as plt # 라이브러리 임포트

plt.title('marker') 

# 'r.--' : <색상><마커모양><선모양> 코드로 작성 가능
plt.plot([10,20,30,40],'r.-',label='circle') 
plt.plot([40,30,20,10],'g^--',label='triangle up') 
# .: 원 / ^: 삼각형 / *: 별
plt.legend() 
plt.show() 
```

데이터 시각화하기

```python
import csv
import matplotlib.pyplot as plt # 1) 라이브러리 임포트

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f) 
header = next(data)

result = [] # 결측치를 제거한 최고 기온 리스트

for row in data:
  if row[-1] != '': # 최고 기온이 ''(빈문자열)이 아니면,
    result.append(float(row[-1]))

result
plt.figure(figsize=(10,2),dpi=200)
plt.plot(result,'r')
plt.show
```
