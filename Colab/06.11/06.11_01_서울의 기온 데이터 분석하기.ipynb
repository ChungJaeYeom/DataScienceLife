```python
import csv

f = open('/content/seoul.csv','r',encoding='cp949') #경로 목적(read 읽기모드, 만약 rw면 읽고 쓰겠다) 인코딩 / 노트플러스 인코딩 따라가면 확인 가능
data = csv.reader(f) #딜리미터(구분자)-> , 로 생략가능 (디폴트 왜? csv 파일이니 기본적으로 , 근데 캐럿이나 ^ (2순위) | 오알(3순위))

#데이터를 한줄씩 출력
for row in data:
  print(row)

f.close()
```

헤더 저장하기

```python
import csv

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f) 

#next가 한번 실행하면 커서를 넘긴다. 다음줄로 넘긴다.
header = next(data)
print(header)
f.close()
```

```python
import csv

f = open('/content/seoul.csv','r',encoding='cp949')
data = csv.reader(f) 

#첫 번째 행을 읽은 후 다음 행으로 이동시키는 함수
header = next(data)
print(header)

print(next(data))
f.close()
```
