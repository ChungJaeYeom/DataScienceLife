step 3) 최고 기온과 최고 기온이었던 날짜 찾기

```python
import csv

max_temp = -999 #최고 기온 값을 저장할 변수 -> 0? -> 0도보다 낮을 수도 있기 때문에 절대 나오지 않을 값 -> 명확한 결과를 위해 극적인 값으로 셋팅
max_date = ''   #최고 기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/seoul.csv','r',encoding='cp949')

data = csv.reader(f) 
header = next(data) # 헤더 추출

for row in data:
  #최고 기온 데이터가 ''(결측치) 확인
  if row[-1] == '':
    row[-1] = -999 # -999를 넣어서 빈 문자열('')이 있던 자리라고 표시
  
  row[-1] = float(row[-1]) # 기온 데이터 실수형으로 변환

  #최고 기온 정보를 저장

  if max_temp < row[-1]:
    max_date = row[0]
    max_temp = row[-1]

f.close()

print('기상 관측 이래 서울의 최고 기온이 가장 높았던 날은', max_date,'로 ', max_temp, '도 였습니다.')
```

step 4) 최저 기온과 최저 기온이었던 날짜 찾기 (스스로 해보기)

```python
import csv

min_temp = 999  #최저 기온 값을 저장할 변수 -> 0? -> 0도보다 높을 수도 있기 때문에 절대 나오지 않을 값 -> 명확한 결과를 위해 극적인 값으로 셋팅
min_date = ''   #최저 기온이 가장 높았던 날짜를 저장하는 변수

f = open('/content/seoul.csv','r',encoding='cp949')

data = csv.reader(f) 
header = next(data) # 헤더 추출

for row in data:
  #최저 기온 데이터가 ''(결측치) 확인
  if row[-2] == '':
    row[-2] = 999 # 999를 넣어서 빈 문자열('')이 있던 자리라고 표시
  
  row[-2] = float(row[-2]) # 기온 데이터 실수형으로 변환

  #최저 기온 정보를 저장

  if min_temp > row[-2]:
    min_date = row[0]
    min_temp = row[-2]

f.close()

print('기상 관측 이래 서울의 최저 기온이 가장 낮았던 날은', min_date,'로 ', min_temp, '도 였습니다.')
```
