```python
df_user = pd.read_csv('/content/광주광역시_시내버스 정류소별 이용객수(카드)_편집.csv',encoding='cp949')
df_user.head()
```

```python
dfs = df_user.sort_values(by=['total'], ascending=False)
```

```python
dfs
```

```python
dfs["total_min"] = dfs["total"] * 0.00001
```

```python
dfs
```

```python
fig = go.Figure(go.Bar(
    x = dfs['연번'],
    y = dfs['total_min'],
))
fig.update_layout(title_text='날짜별 기타 막대그래프')
fig.show()
```

```python
df_user["비교"] = 0
```

```python
df_user.loc[144, 'total']
```

```python
for n in df_user.index:
  if df_user.loc[n, 'total'] >= df_user.loc[144, 'total']:
    df_user.loc[n, '비교'] = 'More'
  else:
    df_user.loc[n, '비교'] = 'Fewer'
```

```python
df_user
```

```python
fig = px.pie(df_user,
             names='비교',
             title='학동중심사입구역 기준 광주버스 이용객 비교')
fig.show()
```

```python
plt.figure(dpi=200)
plt.title("학동중심사입구역 기준 광주버스 이용객 비교")
explode = [0,0.1]
colors = ['silver','gold']
df_user.비교.value_counts().plot.pie(explode=explode,colors=colors, autopct='%.1f%%')
plt.legend()

plt.show()
```
