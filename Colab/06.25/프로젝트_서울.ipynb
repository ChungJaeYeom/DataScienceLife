```python
import csv
import matplotlib.pyplot as plt

f = open('/content/재개발.csv', encoding = 'cp949')
data = csv.reader(f)

y2019 = []
y2018 = []

for row in data:
  if '2019' in row[0]:
    for i in range(0,18):
      y2019.append(int(row[i+1]))
  
  if '2018' in row[0]:
    for i in range(0,18):
      y2018.append(-int(row[i+1]))

print(y2019)
print(y2018)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=110) 
plt.rc('font',family='NanumBarunGothic')

city = ['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']

plt.barh(range(18), y2019, color = 'skyblue',label = '2019')
plt.barh(range(18), y2018, color = 'hotpink', label = '2018')
plt.yticks(range(len(city)), city)
plt.title('지역별 재개발 & 재건축 현황')
plt.legend()
plt.show()
```

```python
f = open('/content/지역별사고발생건수.csv', encoding = 'cp949')
data = csv.reader(f)

danger = []
city = ['서울','부산','대구','인천','광주','대전','울산','세종','경기','강원','충북','충남','전북','전남','경북','경남','제주']

for row in data:
  if '붕괴' in row[0]:
    for i in range(0,17):
      danger.append(int(row[i+1]))

plt.title('지역별 붕괴사고 발생현황')
plt.bar(range(len(danger)),danger,color = 'lightslategrey',label='붕괴사고 발생건수')
plt.xticks(range(len(city)), city)
plt.legend()
plt.show()
```

```python
f = open('/content/시간별 사망재해_(건설).csv', encoding='cp949')
data = csv.reader(f)
next(data) #헤더

y2016 = [] 
y2017 = []
y2018 = []
time = ['0~2','2~4','4~6','6~8','8~10','10~12','12~14','14~16','16~18','18~20','20~22','22~24']

for row in data:
  if '2016' in row[0]:
    for i in range(12):
      y2016.append(int(row[i+1]))
  if '2017' in row[0]:
    for i in range(12):
      y2017.append(int(row[i+1]))
  if '2018' in row[0]:
    for i in range(12):
      y2018.append(int(row[i+1]))

print(y2016)
print(y2017)
print(y2018)
```

```python
import matplotlib.pyplot as plt

plt.figure(dpi=110) 
plt.rc('font',family='NanumBarunGothic')

plt.plot(y2016,'o', color = 'Deeppink',label = '2016')
plt.plot( y2017,'o', color = 'crimson',label = '2017')
plt.plot(y2018, 'o',color = 'OrangeRed', label = '2018')

plt.xticks(range(len(time)), time)
plt.title('시간별 건설업 사망재해 현황')
plt.legend()
plt.show()
```
